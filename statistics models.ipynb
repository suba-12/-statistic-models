# 2. Pie diagram

```python
# simple pie plot

students = ["harish","suba","hema","prtahibha"]
val  = [34,37,67,87]
plt.pie(val,labels = students)
plt.show()
```

```python
#explode
exp = [0.2,0,0,0]
plt.pie(val,labels=students,startangle = 90,explode=exp)
plt.show()
```

```python
#Autopct
plt.pie(val,labels = students,startangle = 90,explode=exp,autopct = '%2.1f%%')
plt.show()
```

# 4. Standard Deviation

pstdev() method calculates the standard deviation from entire population

```python
# Without PSTDEV()

test = [4,5,8,6,9]
print("the original list :",test)
mean = sum(test)/len(test)
var = sum([((x-mean)** 2)for x in test])/len(test)
res = var**0.5
print("standard deviation is ",res)

```

```python
#with PSTDEV()

print(statistics.pstdev(test))
```

# 7. COMPUTATION RANK CORRELATION

```python

import scipy.stats as cor
```

```python
X = [45,15,78,12]
Y = [45,75,62,76]
```

```python
def corr_ans(X,Y):
    X_rank = pd.Series(X).rank()
    print(X_rank)
    Y_rank = pd.Series(Y).rank()
    print(Y_rank)
    print("Rank of correlation is :",cor.pearsonr(X_rank,Y_rank)[0])
    
```

```python
corr_ans(X,Y)
```

# 8. Regression Line

```python
from scipy import stats
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope,intercept, r,p,s = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
